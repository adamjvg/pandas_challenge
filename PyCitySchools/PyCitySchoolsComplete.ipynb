```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Renaming the columns to make them a little easier to read
school_data_complete = school_data_complete.rename(columns={"student_name": "Student Name",
                                        "gender": "Gender",
                                        "grade": "Grade",
                                        "reading_score": "Reading Score",
                                        "math_score" : "Math Score",
                                        "type" : "Type",
                                        "size" : "Size",
                                        "budget": "Budget"
                                                           })

school_data_complete.head()
```

```python
#calculate total number of schools
total_schools = len(school_data_complete["School ID"].unique())

total_schools
```

```python
#Calculate the total number of students
total_students = len(school_data_complete["Student ID"].unique())

total_students
```

```python
#Calculate the total budget
budget = school_data["budget"].sum()

budget
```

```python
#Calculate average math score
avg_math_score = school_data_complete["Math Score"].mean()

avg_math_score
```

```python
#Calculate average reading score
avg_reading_score = school_data_complete["Reading Score"].mean()

avg_reading_score
```

```python
#Calculate overall pass rate
avg_overall_score = (avg_reading_score + avg_math_score)/2

avg_overall_score
```

```python
#Percentage of students with a score over 70% in maths(a pass)
school_data_complete["Passing Math"] = school_data_complete["Math Score"] >= 70
passing_math = ((school_data_complete["Passing Math"]).mean())*100

passing_math
```

```python
school_data_complete["Passing Reading"] = school_data_complete["Reading Score"] >= 70
passing_reading = ((school_data_complete["Passing Reading"]).mean())*100

passing_reading
```

```python
#overall passing rate
passing_rate = (passing_reading + passing_math)/2

passing_rate
```

```python
#put all new variables in a dataframe

district_results = [{"Total Schools":total_schools,
                       "Total Students":total_students,
                       "Total Budget":budget,
                       "Average Math Score":round(avg_math_score,2),
                       "Average Reading Score":round(avg_reading_score,2),
                       "Average Overall Score":round(avg_overall_score,2),
                       "% Passing Math": passing_math,
                       "% Passing Reading": passing_reading,
                       "% Overall Passing Rate":passing_rate,
                       }]

district_summary_df = pd.DataFrame(district_results)

#clean it up with formatting
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:,.2f}%".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:,.2f}%".format)
district_summary_df["% Overall Passing Rate"] = district_summary_df["% Overall Passing Rate"].map("{:,.2f}%".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#find out whether students are passing maths and reading in a new column
school_data_complete["Passing Math"] = school_data_complete["Math Score"] >=70
school_data_complete["Passing Reading"] = school_data_complete["Reading Score"] >=70

school_data_complete.head()
```

```python
#Group by school name
school_group_df = school_data_complete.groupby(["school_name"]).mean()
school_group_df["Per Student Budget"] = school_group_df["Budget"]/school_group_df["Size"]
school_group_df["Passing Math %"] = round(school_group_df["Passing Math"]*100,2)
school_group_df["Passing Reading %"]= round(school_group_df["Passing Reading"]*100,2)
school_group_df["Passing Overall %"]= round(((school_group_df["Passing Math"] + school_group_df["Passing Reading"])/2)*100,2)

school_group_df.head()
```

```python
#Merge with original school_data dataframe #drop everything but type, size and budget
school_summary = pd.merge(school_group_df, school_data, how="left", on=["school_name","school_name"])

#drop unnecessary columns
del school_summary["size"]
del school_summary["School ID_x"]
del school_summary["Student ID"]
del school_summary["budget"]
school_summary.head()
```

```python
#create a new Dataframe for results
school_summary_df = pd.DataFrame({"School Name":school_summary["school_name"],
                                "School Type": school_summary["type"],
                                 "Total Students": school_summary["Size"],
                                 "Total School Budget": school_summary["Budget"],
                                 "Budget Per Student": school_summary["Per Student Budget"],
                                 "Average Math Score" :round(school_summary["Math Score"],2),
                                 "Average Reading Score" :round(school_summary["Reading Score"],2),
                                 "Passing Math %":school_summary["Passing Math %"],
                                 "Passing Reading %":school_summary["Passing Reading %"],
                                 "Passing Overall %":school_summary["Passing Overall %"]})

#format
school_summary_df["Total Students"] = school_summary_df["Total Students"].map("{:,.0f}".format)
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)

school_summary_df
```
